```python
#Memorization program
def water_jug_problem(jug1_cap, jug2_cap, target):
    visited = set()
    path = []
    def dfs(jug1, jug2):
        if (jug1, jug2) in visited:
            return False
        visited.add((jug1, jug2))
        path.append((jug1, jug2))
        if jug1 == target or jug2 == target:
            return True
        actions = [
            (jug1_cap, jug2),
            (jug1, jug2_cap),
            (0, jug2),
            (jug1, 0),
            (jug1 - min(jug1, jug2_cap - jug2),
             jug2 + min(jug1, jug2_cap - jug2)),
            (jug1 + min(jug2, jug1_cap - jug1),
             jug2 - min(jug2, jug1_cap - jug1))
        ]
        for next_state in actions:
            if dfs(*next_state):
                return True
        path.pop()
        return False
    if dfs(0, 0):
        for step, state in enumerate(path):
            print(f"Step {step}: Jug1 = {state[0]}L, Jug2 = {state[1]}L")
    else:
        print("No solution found.")
jug1 = int(input("Enter capacity of Jug 1: "))
jug2 = int(input("Enter capacity of Jug 2: "))
target = int(input("Enter target amount: "))
water_jug_problem(jug1, jug2, target)
```

```python
#DFS program
def water_jug_dfs(jug1_cap, jug2_cap, target):
    visited = set()
    path = []
    def dfs(jug1, jug2):
        if (jug1, jug2) in visited:
            return False
        visited.add((jug1, jug2))
        path.append((jug1, jug2))
        if jug1 == target or jug2 == target:
            return True
        next_states = [
            (jug1_cap, jug2),
            (jug1, jug2_cap),
            (0, jug2),
            (jug1, 0),
            (jug1 - min(jug1, jug2_cap - jug2),
             jug2 + min(jug1, jug2_cap - jug2)),
            (jug1 + min(jug2, jug1_cap - jug1),
             jug2 - min(jug2, jug1_cap - jug1))
        ]
        for state in next_states:
            if dfs(state[0], state[1]):
                return True
        path.pop()
        return False
    if dfs(0, 0):
        for i, (a, b) in enumerate(path):
            print(f"Step {i}: Jug1 = {a}L, Jug2 = {b}L")
    else:
        print("No solution found.")
jug1 = int(input("Enter capacity of Jug 1: "))
jug2 = int(input("Enter capacity of Jug 2: "))
target = int(input("Enter target amount: "))
water_jug_dfs(jug1, jug2, target)
```

```python
#BFS program
from collections import deque

def water_jug_bfs(jug1_cap, jug2_cap, target):
    visited = set()
    queue = deque()
    parent = {}
    start = (0, 0)
    queue.append(start)
    visited.add(start)
    parent[start] = None
    while queue:
        jug1, jug2 = queue.popleft()
        if jug1 == target or jug2 == target:
            path = []
            state = (jug1, jug2)
            while state:
                path.append(state)
                state = parent[state]
            path.reverse()
            for i, (a, b) in enumerate(path):
                print(f"Step {i}: Jug1 = {a}L, Jug2 = {b}L")
            return
        next_states = [
            (jug1_cap, jug2),
            (jug1, jug2_cap),
            (0, jug2),
            (jug1, 0),
            (jug1 - min(jug1, jug2_cap - jug2),
             jug2 + min(jug1, jug2_cap - jug2)),
            (jug1 + min(jug2, jug1_cap - jug1),
             jug2 - min(jug2, jug1_cap - jug1))
        ]
        for state in next_states:
            if state not in visited:
                visited.add(state)
                parent[state] = (jug1, jug2)
                queue.append(state)

    print("No solution found.")
jug1 = int(input("Enter capacity of Jug 1: "))
jug2 = int(input("Enter capacity of Jug 2: "))
target = int(input("Enter target amount: "))
water_jug_bfs(jug1, jug2, target)
```
