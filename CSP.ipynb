```python
from itertools import permutations

n = int(input("Enter number of words to sum: "))
words = [input(f"Word {i+1}: ").upper() for i in range(n)]
result = input("Result word: ").upper()
letters = set(''.join(words) + result)
letters = list(letters)

if len(letters) > 10:
    print("Too many letters! Maximum 10 letters for digits 0-9.")
    exit()
leading_letters = set(word[0] for word in words + [result])
digits = range(10)
def is_valid(mapping):
    for letter in leading_letters:
        if mapping[letter] == 0:
            return False
    word_values = []
    for word in words:
        num = int(''.join(str(mapping[c]) for c in word))
        word_values.append(num)
    res_value = int(''.join(str(mapping[c]) for c in result))
    return sum(word_values) == res_value
found_solution = False
for perm in permutations(digits, len(letters)):
    mapping = dict(zip(letters, perm))
    if is_valid(mapping):
        print("\nSolution found:")
        for letter in letters:
            print(f"{letter} = {mapping[letter]}")
        for word in words:
            print(''.join(str(mapping[c]) for c in word))
        print('_' * max(len(result), max(len(w) for w in words)))
        print(''.join(str(mapping[c]) for c in result))
        found_solution = True
        break
if not found_solution:
    print("No solution exists.")
```
